```python
M = 19
S = 4
q = []
qt = []
jkr = []
def jk(a, b):
    s = a + b
    if s == "00": return "0x"
    if s == "01": return "1x"
    if s == "10": return "x1"
    if s == "11": return "x0"

for i in range(M):
    q.append(bin(i)[2:].rjust(5, '0'))
    qt.append(bin((i + S)%M)[2:].rjust(5, '0'))
    jkr.append([jk(q[i][j], qt[i][j]) for j in range(5)])
    print(i, bin(i)[2:].rjust(5, '0'), (i + S)%M, bin((i + S)%M)[2:].rjust(5, '0'), *jkr[i])
    
```

```python
counter = 4 # D Триггер
for elem in list(zip(*qt)):
    f = "".join(elem) # .ljust(32, 'x')
    print
    # print(f)
    # print(len(f))
    rows = ['00', '01', '11', '10']
    cols = ['000', '001', '011', '010', '110', '111', '101', '100']
    arr = [['x' for __ in range(len(cols))] for _ in range(len(rows))]
    for i in range(M):
        s = q[i]
        arr[rows.index(s[:2])][cols.index(s[2:])] = f[i]
    print(f"Q:{counter}")
    counter -= 1
    for i in range(4):
        print("".join(arr[i]))
    print("----------------")
```
