# Первый эксперимент

```python
import numpy as np
f1 = [i*10 for i in range(1, 6)]
f1 = [5] + f1
f1 = np.array(f1)
dt = np.array([15.85, 13.14, 9.06, 6.75, 5.35, 4.42])
phi = 360*f1*dt/1000
print(phi)
```

```python
L = 1.94
Xl = 2*np.pi*f1*L
print(Xl)
```

```python
print(f1)
```

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots()
ax.plot(f1, phi)
```

```python
fig, ax = plt.subplots()
ax.plot(f1, Xl)
```

# 3-й Эксперимент

```python
fr = 19.91
arr = []
for i in range(3, 11, 2):
    arr.append(fr*i/10)
arr.append(fr)
arr += [1.1*fr, 1.2*fr, 1.5*fr, 2*fr]
```

```python
dt = np.array([-37.74, -20.22, -11.29, -3.49, 0.305, 3.01, 4.58, 5.7, 5.09])
I = np.array([393.82, 769.68, 1400, 2360, 2590, 2400, 2003, 1260, 767.97])
Ul = np.array([28.66, 93.34, 237.38, 515.81, 629.04, 639.42, 592.13, 459.64, 372.76])
Uc = np.array([318.15, 373.02, 484.02, 636.24, 627.86, 527.5, 410.5, 203.83, 93.01])
Ur = np.array([44.5, 86.96, 157.97, 266.99, 292.89, 270.67, 229.77, 142.65, 86.78])
U = 292.89
f = np.array(arr)
phi = 360*f*dt/1000
Z = U*1000/I
print(arr, phi, Z,sep="\n")
```

```python
fig, ax = plt.subplots()
ax.plot(f, Ul, label="Ul")
ax.plot(f, Uc, label="Uc")
ax.plot(f, Ur, label="Ur")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(f, I)
```

```python
fig, ax = plt.subplots()
ax.plot(f, phi)
```

```python
fig, ax = plt.subplots()
ax.plot(f, Z)
```
