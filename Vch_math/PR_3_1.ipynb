```python
import numpy as np
import matplotlib.pyplot as plt 
from math import atan, pi

def f1(x):
    return pi/2 - atan(x) + 2*x - 1
def f2(x):
    return 3*x**4 + 4*x**3 - 12*x**2 + 1

def f1_np(x):
    return pi/2 - np.arctan(x) + 2*x - 1
def f2_np(x):
    return 3*np.power(x, 4) + 4*np.power(x, 3) - 12*np.power(x, 2) + 1

e = 0.01

```

```python
a = -2
b = 0
x_prev = None
x = None
print("a b x f b - a")
while 1:
    x_mid = (b + a)/2
    
    f = f1(x_mid)
    print(a, b, x_mid,  f, b-a)
    if f > 0:
        b = x_mid
    elif f < 0:
        a = x_mid
    else:
        print("answer: ", x_mid)
        break
    if x_prev is None:
        x_prev = x_mid
    else:
        if abs(x_prev - x_mid) <= 0.01:
            print("x: ", x_mid)
            x = x_mid
            break
    x_prev = x_mid
```
