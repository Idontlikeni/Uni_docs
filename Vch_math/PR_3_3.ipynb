```python
from math import *

e = 0.001

def f1(x):
    return 2*x - log10(x) - 7

def f2(x):
    return x ** 3 - 3*x**2 + 6*x - 5

def f1d(x):
    return 2 - 1/(x*log(10, e))

def f2d(x):
    return 3*x ** 2 - 6*x + 6
```

```python
x = f1(6)
while 1:
    temp_x = x - f1(x)/f1d(x)
    print(temp_x)
    if abs(temp_x - x) <= e:
        print("x: ", temp_x)
        break
    x = temp_x
```

```python
x = f2(1)
while 1:
    temp_x = x - f2(x)/f2d(x)
    print(temp_x)
    if abs(temp_x - x) <= e:
        print("x: ", temp_x)
        break
    x = temp_x
```
