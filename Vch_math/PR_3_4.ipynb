найти правый и левый, потом сложить и поделить на 2
$$ \text{Интеграл} = \frac{(a-b)}{n}\sum_{i=0}^{n-1}(x_{i})$$

```python
import numpy as np
def f1(x):
    return np.sqrt(x*x + 0.5) / (2*x + np.sqrt(x*x + 2.5))

def f2(x):
    return np.cos(x*x + 1) / (2 + np.sin(2*x + 0.5))
```

```python
h = (1.4 - 0.6)/9
h
```

```python
x = np.array([0.6 + i * h for i in range(10)])
print(*x, sep='\n')
```

```python
y = f1(x)
print(*y, sep='\n')
```

```python
smy1 = np.sum(y[:-1])
smy2 = np.sum(y[1:])
smy1, smy2
```

```python
I1 = h * smy1
I2 = h * smy2
I3 = (I1 + I2) / 2
I1, I2, I3
```

Проверка итеграла:
