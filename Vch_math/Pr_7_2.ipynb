# Задание 7.2
Найти минимум функции $4x^{2}+y^{2}-4x-2y$ двух переменных методом градиентного спуска
используя начальное приближение $(1, 0)$ с шагом α=0.2, и точностью
ε=0.1:

```python
import numpy as np
def grad(x, y): # градиент
    return [8*x - 4, 2*y - 2]
```

```python
a = 0.2
e = 0.1
```

```python
x = np.array(grad(1, 0))
while True:
    g = a*np.array(grad(*x))
    print("Градиент: ", g)
    print("норма градиента: ", (g[0]**2 + g[1]**2)**0.5)
    if((g[0]**2 + g[1]**2)**0.5) <= e:
        break
    x = x - g
    print("X: ", x)

print(x)
```
